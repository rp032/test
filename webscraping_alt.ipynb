```python
# Mit Request 

url = 'https://www.marc-cain.com/Outlet/Kleider/'
html = requests.get(url)
```

```python
# mit einer Response 200 wissen wir, dass wir die Seite scrappen können -> Response war erfolgreich 

html
```

```python
# In soup findet man die relevanten Daten einer Seite 

soup = BeautifulSoup(html.text, 'html.parser')
```

```python
originalpreis = soup.find_all('span', {'class': 'product-box__details__price__original'})
originalpreis
```

```python
a =0
originalpreisliste= []
while a<10:
    originalpreis = soup.find_all("span", class_="product-box__details__price__original") [a]
    originalpreis = originalpreis.text.replace ('â\x82¬',"")
    a = a+1
    originalpreisliste.append(originalpreis)
       
```

```python
originalpreisliste
```

```python
b =0
salepreisliste= []
while b<10:
    salepreis = soup.find_all("span", class_="product-box__details__price__sale") [b]
    salepreis = salepreis.text.replace ('â\x82¬',"")
    b = b+1
    salepreisliste.append(salepreis)

   
       
       



    
```

```python
salepreisliste
```

```python
name = soup.find_all('div', {'class': 'product-box__details__name'})
name
```

```python
c =0
nameliste= []
while c<10:
    name = soup.find_all("div", class_="product-box__details__name") [c]
    name = name.text.replace ("\n", "")
    c = c+1
    nameliste.append(name)



```

```python
nameliste
```

```python
df = pd.DataFrame (
    {"originalpreis":originalpreisliste,
    "salepreis":salepreisliste,
    "porduktname":nameliste})
```
