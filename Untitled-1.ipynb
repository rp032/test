Aufgabe D

Hypothese: Die CPC ist auf Facebook höher 
Wahl des Hypothesentest: abhängiger T-Test, da wir den gleichen Datensatz/Dataframe verwenden

```python
import numpy as np
from scipy import stats
```

```python
df_fb = df_fb[(df_fb["CPC (Kosten pro Link-Klick) (EUR)"].notnull())]
df_insta = df_insta[(df_insta["CPC (Kosten pro Link-Klick) (EUR)"].notnull())]

GruppeFB = df_fb["CPC (Kosten pro Link-Klick) (EUR)"][0:32]
GruppeInsta = df_insta ["CPC (Kosten pro Link-Klick) (EUR)"][0:32]

```

```python
GruppeInsta =GruppeInsta.astype(int)
GruppeFB = GruppeFB.astype(int)
```

```python
np.random.seed(123)
```

```python
dftest = pd.DataFrame({'GruppeInsta': GruppeInsta, "GruppeFB": GruppeFB})
```

```python
%matplotlib inline
sns.set_style("whitegrid")

sns.histplot(x=GruppeFB, data=df_fb, color="blue", label="GruppeFB", kde=True)
sns.histplot(x=GruppeInsta, data=df_insta, color="red", label="GruppeInsta", kde=True, alpha=0.8);
```

```python
t, p = stats.ttest_ind(GruppeInsta, GruppeFB)
print("t =", t)
print("p =", p)

# P ist > 0,05 deswegen nehmen wir die 0 Hypothese an
```

```python
diff_mean = GruppeInsta.mean() - GruppeFB.mean()
print(round(diff_mean,2))
```
